```python
import wooldridge as woo
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = woo.data("wage1")
```

```python
fig,axs = plt.subplots(2,2,figsize=(10,10), tight_layout=True,
                       gridspec_kw={'height_ratios': [1,4],'width_ratios': [4,1]})

sns.scatterplot(x="educ",y="wage",hue="female",data=df,ax=axs[1,0])
sns.kdeplot(x="educ",hue="female",data=df,ax=axs[0,0], fill=True)
sns.kdeplot(y="wage",hue="female",data=df,ax=axs[1,1], fill=True)

axs[0,0].axis("off")
axs[0,1].axis("off")
axs[1,1].axis("off")

axs[0,0].get_legend().remove()
axs[1,1].get_legend().remove()

plt.show()
```
